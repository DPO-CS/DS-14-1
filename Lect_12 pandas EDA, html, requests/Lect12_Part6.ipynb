### Cлучайный вопрос из [базы вопросов ЧГК](https://db.chgk.info/)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://db.chgk.info/tour/mkm18_u' # возьмем конкретный турнир

page = requests.get(url)
page.encoding = 'utf-8'
soup = BeautifulSoup(page.text)

print(soup.prettify())
```

```python
soup.find_all("div", {"class": "question"})[0]
```

```python
questions = soup.find_all("div", {"class": "question"})
num_questions = len(questions)
num_questions
```

```python
questions[0].find_all('p')[1].text.strip()
```
